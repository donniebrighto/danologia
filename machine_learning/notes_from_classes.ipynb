```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Pierwszy webscraping (drobny, ale zawsze)
import urllib.request
import os

filename = 'pomiary.xlsx'
url = 'https://modeleisystemy.pl/wp-content/uploads/2017/03/' + filename
if not os.path.isfile(filename):
    print('Pobieranie... ' + url)
    urllib.request.urlretrieve(url, filename)
    
pomiary = pd.read_excel(filename)
display(pomiary)
print(pomiary.head())
print(pomiary.tail())
```

```python
pomiary.mean() # parametr axis = 0 lub 1
```

```python
# pomiary.drop(parametry) <- usuwanie jakichś danych
# zamiast usuwania rekordów niekompletnych wypełniamy dane fałszywe np. zerami lub średnią

pomiary = pomiary.fillna(pomiary.mean(axis = 0))
display(pomiary.head())
```

```python
# zamiana błednych danych
waga = pomiary.loc[1, 'wzrost']
pomiary.loc[1, 'wzrost'] = pomiary.loc[1, 'waga']
pomiary.loc[1, 'waga'] = waga
display(pomiary)
```

```python
pomiary.keys().shape
```

```python
osx = 'waga'
osy = 'wzrost'

X = pomiary[osx]
Y = pomiary[osy]

print(X.shape)
print(Y.shape)

# (43,1) <- tabela która ma dwa wymiary, pierwszy 43, drugi 1
# (43, ) <- tabela z jednym wymiarem
# python -> funkcja reshape - przetwarzanie jednego wymiaru w drugi

plt.scatter(X, Y, marker = 'o', facecolor = 'white', edgecolors = 'black')
plt.xlabel(osx, fontsize = 16)
plt.ylabel(osy, fontsize = 16)
plt.show()
```

```python
# wyrzucanie outlier'ów - danych odstających
temp = pomiary[pomiary[osx] < 120]

X = temp[osx]
Y = temp[osy]

print(X.shape)
print(Y.shape)

# (43,1) <- tabela która ma dwa wymiary, pierwszy 43, drugi 1
# (43, ) <- tabela z jednym wymiarem
# python -> funkcja reshape - przetwarzanie jednego wymiaru w drugi

plt.scatter(X, Y, marker = 'o', facecolor = 'white', edgecolors = 'black')
plt.xlabel(osx, fontsize = 16)
plt.ylabel(osy, fontsize = 16)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

X = X.reshape(-1, 1) # pierwszy wymuszamy, a drugi jest dobierany 'z automatu'
Y = Y.reshape(-1, 1)

print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X, Y, marker = 'o', facecolor = 'white', edgecolors = 'black')

x_tr, x_ts, y_tr, y_ts = train_test_split(X, Y)

# REGRESJA LINIOWA
linreg = LinearRegression().fit(x_tr, y_tr)
plt.scatter(X, linreg.predict(X), color='red')
print("Zbiór treningowy: ", linreg.score(x_tr, y_tr))
print("Zbiór testujący: ", linreg.score(x_ts, y_ts))

# SVM
svmreg = SVR(gamma='auto').fit(x_tr, y_tr.ravel())
# ravel i flatten <- spłaszczanie (43, 1) do (43, )
# https://stackoverflow.com/questions/28930465/what-is-the-difference-between-flatten-and-ravel-functions-in-numpy
plt.scatter(X, svmreg.predict(X), color='green')

plt.xlabel(osx, fontsize = 16)
plt.ylabel(osy, fontsize = 16)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_tr_pred = linreg.predict(x_tr) # y = a * x + b
y_ts_pred = linreg.predict(x_ts)

print("Błąd dla zbioru treningowego: ", mean_squared_error(y_tr, y_tr_pred))
print("Błąd dla zbioru testowego: ", mean_squared_error(y_ts, y_ts_pred))
```

```python
X = np.array(temp[osx])
Y = np.array(temp[osy])

dane = np.array([X, Y]).T

import sklearn.cluster

km = sklearn.cluster.KMeans(n_clusters=2)
km.fit(dane)

plec_pred = km.predict(dane)
plec_pred = ['grey' if p == 0 else 'red' for p in plec_pred]
plt.scatter(X, Y, marker='o', facecolors=plec_pred)
plt.show()
```

```python
import sklearn.decomposition as skld

pca = skld.PCA(n_components=2)
pca.fit(dane)
plec_pred = km.predict(dane)
dane2 = pca.transform(dane)
gr0 = (plec_pred == 0)
plt.scatter(dane2[gr0, 0], dane2[gr0, 1], marker = 'o', facecolor = 'red')
plt.scatter(dane2[~gr0, 0], dane2[~gr0, 1], marker = 'o', facecolor = 'grey')
plt.xlabel(osx, fontsize = 16)
plt.ylabel(osy, fontsize = 16)
plt.show()
```
